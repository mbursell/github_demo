```python
import pandas as pd
```

```python
%matplotlib inline
import plotnine as p9
```

```python
surveys_df = pd.read_csv('surveys.csv')
surveys_df.head()
```

```python
surveys_complete = surveys_df.dropna()
```

```python
plot = (p9.ggplot(data=surveys_complete,
           mapping=p9.aes(x='weight',
                          y='hindfoot_length',
                          color='species_id'))
    + p9.geom_point(alpha=0.1)
)
plot
```

```python
plot.save("scatterplot.png", width=10, height=10, dpi=300)
```
